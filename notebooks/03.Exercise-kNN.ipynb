```python
from sklearn.datasets import load_wine
from sklearn import model_selection
from sklearn.metrics import confusion_matrix, classification_report 
import pandas as pd

# Load the wine dataset
wine_data = load_wine()

# Convert to DataFrame for easier handling
wine_df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

# Add the target labels to the DataFrame
wine_df['target'] = wine_data.target

# Display the first few rows
print(wine_df.head())
```

## Task

**Your task is to classify this wine data using kNN. Try to predict the target value from the other factors. Don't forget to create a train/test split. You should print a confusion matrix and a classification report**

```python
features = np.hstack([wine_df.values[::2, :-1], wine_df.values[1::2, :2]])
target = wine_df.values[1::2, -1]

X_train, X_test, y_train, y_test = model_selection.train_test_split(
    features, target, test_size=0.1, random_state=42
)

train_data = np.array(X_train, dtype=np.float32) 
labels = np.array(y_train, dtype=np.int32)      

knn = cv2.ml.KNearest_create() 
knn.train(train_data, cv2.ml.ROW_SAMPLE, labels)

test_data = np.array(X_test, dtype=np.float32)

ret, predictions, neighbors, dist = knn.findNearest(test_data, k=3)

print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))

print("\nClassification Report:")
print(classification_report(y_test, predictions,zero_division=0))
```
